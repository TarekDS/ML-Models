```python
def plot_hist(x, p=5):
    # Plot the distribution and mark the mean
    pyplot.hist(x, alpha=.5)
    pyplot.axvline(x.mean())
    # 95% confidence interval    
    pyplot.axvline(numpy.percentile(x, p/2.), color='red', linewidth=3)
    pyplot.axvline(numpy.percentile(x, 100-p/2.), color='red', linewidth=3)
    
def plot_dists(a, b, nbins, a_label='pop_A', b_label='pop_B', p=5):
    # Create a single sequence of bins to be shared across both
    # distribution plots for visualization consistency.
    combined = pandas.concat([a, b])
    breaks = numpy.linspace(
        combined.min(), 
        combined.max(), 
        num=nbins+1)

    pyplot.subplot(2, 1, 1)
    plot_hist(a)
    pyplot.title(a_label)
    
    pyplot.subplot(2, 1, 2)
    plot_hist(b)
    pyplot.title(b_label)
    
    pyplot.tight_layout()
    

plot_dists(below2555.price, above2555.price, 20, a_label='below2555', b_label='above2555')
```

```python
#feature selection
auto1 = auto[['wheel-base','length','width','height', 'engine-size', 'compression-ratio', 'city-mpg','horsepower','peak-rpm', 'highway-mpg', 'price']]
```

```python
import numpy as np
import pandas as pd

auto['cw'] = np.where(auto['curb-weight']>=2555, 'yes', 'no')
```

```python
for col in ['cw']:
    auto[col] = auto[col].astype('category')
```

```python
#categorizing label
label = auto['cw'].cat.codes

#greater then 2555 is set 1 and less then 2555 is set to 0
```

```python
#normalizing features
import pandas as pd
from sklearn import preprocessing

x = auto1.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
auto2 = pd.DataFrame(x_scaled)
```

```python
X = label
y = auto2
```

```python
feature_vecs = auto2.as_matrix() 
```

```python
import sklearn.naive_bayes

# Define the model
model = sklearn.naive_bayes.MultinomialNB(alpha=1e-7)
# Train the model with our votes dataset
model.fit(feature_vecs, label)
```

```python
#prediction vs actual
predicted_cw = model.predict(feature_vecs)
cw_probabilities = model.predict_proba(feature_vecs)

results = pandas.DataFrame({
        'cw': auto['cw'],
        'predicted': pandas.Categorical.from_codes(
            predicted_cw, auto['cw'].cat.categories),
        'proba(below2555)': cw_probabilities[:, 0],
        'proba(above2555)': cw_probabilities[:, 1],
    })
results
```

```python
import sklearn.metrics

def confusion_matrix(labels, predicted_labels, label_classes):
    return pandas.DataFrame(
        sklearn.metrics.confusion_matrix(labels, predicted_labels),
        index=[label_classes], 
        columns=label_classes)    
 
def performance(results):
    accuracy = sklearn.metrics.accuracy_score(
        results['cw'].cat.codes, results['predicted'].cat.codes)
    precision = sklearn.metrics.precision_score(
            results['cw'].cat.codes, results['predicted'].cat.codes)
    recall = sklearn.metrics.recall_score(
            results['cw'].cat.codes, results['predicted'].cat.codes)

    print('Accuracy = %.3f, Precision = %.3f, Recall = %.3f' % (accuracy, precision, recall))
    
    return confusion_matrix(
        results['cw'], 
        results['predicted'], 
        results.cw.cat.categories)
```

```python
performance(results)
```
